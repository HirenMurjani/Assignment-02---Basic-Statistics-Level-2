```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import zscore

companies = ["Allied Signal", "Bankers Trust", "General Mills", "ITT Industries",
             "J.P.Morgan & Co.", "Lehman Brothers", "Marriott", "MCI",
             "Merrill Lynch", "Microsoft", "Morgan Stanley", "Sun Microsystems",
             "Travelers", "US Airways", "Warner-Lambert"]
measure = [24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26,
             32.95, 91.36, 25.99, 39.42, 26.71, 35.00]

plt.figure(figsize=(10, 6))
plt.barh(companies, measure, color='skyblue')
plt.xlabel('Measure X (%)')
plt.title('Measure X for Different Companies')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()

# Calculate mean, standard deviation, and variance
mean_x = np.mean(measure)
stdev_x = np.std(measure)
var_x = np.var(measure)

# Identify outliers using z-score
z_scores = zscore(measure)
outliers = np.where(np.abs(z_scores) > 3)[0]
outlier_companies = [companies[i] for i in outliers]

# Output results
print(f"Mean (): {mean_x:.2f} ")
print(f"Standard Deviation (): {stdev_x:.2f}")
print(f"Variance (^2): {var_x:.2f}")
print(f"Outliers: {outlier_companies}")
```
